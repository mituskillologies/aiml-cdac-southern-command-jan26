```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
list(enumerate(alphabets))
```

```python
int_to_char = dict((i,v) for i,v in enumerate(alphabets))
char_to_int = dict((v,i) for i,v in enumerate(alphabets))
```

```python
char_to_int
```

```python
for i, v in enumerate(alphabets):
    print(v,i)
```

#### Prepare the dataset

```python
seq_length = 1
x = []
y = []

for i in range(len(alphabets) - seq_length): 
    seq_in = [alphabets[i]]
    seq_out = alphabets[i+1] 
    print(seq_in,'-->',seq_out)
    x += [[char_to_int[seq_in[0]]]] 
    y += [char_to_int[seq_out]]
```

```python
x
```

#### Reshape and normalize the input 

```python
import numpy as np
```

```python
x = np.reshape(x,(25,1,1))
```

```python
x;
```

```python
x = x / 25
```

```python
x
```

#### Prepare the output

```python
from keras.utils import to_categorical
```

```python
y_new = to_categorical(y)
```

```python
y_new.shape
```

#### Build the model

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Input
```

```python
model = Sequential()

# input layer
model.add(Input((1,1)))

# LSTM Layer
model.add(Dense(25, activation='relu'))
model.add(LSTM(32))

# output layer
model.add(Dense(26, activation='softmax'))
```

```python
model.summary()
```

#### Compile the model

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', 
             metrics=['accuracy'])
```

#### Train the model

```python
x.shape
```

```python
y_new.shape
```

```python
model.fit(x, y_new, epochs=500, batch_size=1)
```

#### Prediction

```python
new = 'P'
new = char_to_int[new] 
new = np.reshape(new, (1,seq_length,1))
new = new / 25
result = model.predict(new, verbose=False)
int_to_char[result.argmax()]
```
