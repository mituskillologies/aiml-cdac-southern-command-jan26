```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.histplot(iris['petal_length']);
```

```python
sns.displot(iris['sepal_width']);
```

```python
sns.histplot(iris['sepal_width'], kde=True);
```

```python
sns.kdeplot(iris['sepal_width']);
```

```python
sns.jointplot(x='petal_length', y = 'sepal_width', data= iris)
```

```python
sns.jointplot(x='petal_length', y = 'sepal_width', hue='species', data= iris)
```

```python
sns.pairplot(iris, hue='species')
```
