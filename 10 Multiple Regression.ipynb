#### Import the libaries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

#### Load the data

```python
df = pd.read_csv('mtcars.csv')
```

```python
df
```

```python
#### Separate the input and output variables
```

```python
# input -> disp, hp, wt 
x = df[['disp','hp','wt']]

# output -> mpg
y = df['mpg']
```

```python
x
```

#### Build the model

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x, y)
```

```python
regressor.score(x, y)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
disp = 221
hp = 102
wt = 2.91
```

```python
disp * -9.37009081e-04 + hp * -3.11565508e-02 + wt * -3.80089058 + 37.105
```

```python
regressor.predict([[disp, hp, wt]])
```

```python
pred = regressor.predict(x)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y, pred)
```
