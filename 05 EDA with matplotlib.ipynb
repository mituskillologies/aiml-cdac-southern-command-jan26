```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

#### Line Plot

```python
x = np.array([2,7,12,16,23,28,34,32,45,51])
y = np.array([10,12,23,33,45,48,61,76,82,85])
z = np.array([77,65,54,57,54,51,34,29,27,27])
p = np.array([56,54,34,25,57,65,76,34,26,55])
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, z)
```

```python
plt.plot(x, p)
```

```python
plt.title("My first plot")
plt.xlabel("X-Value")
plt.ylabel("Y-Value")
plt.grid()
plt.plot(x, y)
```

```python
plt.title("My first plot")
plt.xlabel("X-Value")
plt.ylabel("Y-Value")
plt.xlim(0,60)
plt.ylim(0,90)
plt.xticks(range(0,61,5))
plt.yticks(range(0,91,5))
plt.grid()
plt.plot(x, y)
```

```python
plt.title("My first plot")
plt.xlabel("X-Value")
plt.ylabel("Y-Value")
plt.grid()
plt.plot(x, y, color = '#67A580', marker = 'o', ls='--', ms = 10, 
        mfc = 'red', lw = 2)
# ls -> line style     --   :   -. 
# ms -> marker size 
# mfc -> marker face color 
# lw -> line width 
# markers -> o . * D > < v ^ 
```

```python
plt.title("Comparison plot")
plt.xlabel("X-Value")
plt.ylabel("Y-Value")
plt.grid()
plt.plot(x, y, color = 'r', marker = '.')
plt.plot(x, z, color = 'g', marker = '.')
```

```python
plt.title("Comparison plot")
plt.xlabel("X-Value")
plt.ylabel("Y-Value")
plt.grid()
plt.plot(x, y, color = 'r', marker = '.', label = '2024')
plt.plot(x, z, color = 'g', marker = '.', label = '2025')
plt.legend(loc='upper left')
```

```python
plt.figure(figsize=(16,9))
plt.title("Comparison plot")
plt.xlabel("X-Value")
plt.ylabel("Y-Value")
plt.grid()
plt.plot(x, y, color = 'r', marker = '.', label = '2024')
plt.plot(x, z, color = 'g', marker = '.', label = '2025')
plt.legend(loc='upper left')
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title("Comparison plot")
plt.xlabel("X-Value")
plt.ylabel("Y-Value")
plt.grid()
plt.plot(x, y, color = 'magenta', marker = 'o', ls = ":")
plt.subplot(1,2,2)
plt.plot(x, z)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.plot(x, y, color = 'magenta', marker = 'o', ls = ":")
plt.subplot(2,2,2)
plt.plot(x, z)
plt.subplot(2,2,3)
plt.plot(y, z)
plt.subplot(2,2,4)
plt.plot(x, p)
```

```python
plt.subplot(2,2,1)
plt.plot(x, y)
plt.subplot(2,2,3)
plt.plot(x, z)
plt.subplot(1,2,2)
plt.plot(x, p)
plt.savefig('myplot.png')
```

#### Bar Plot

```python
df = pd.read_csv('student-info.csv', nrows=10)
```

```python
df.shape
```

```python
plt.figure(figsize=(16,9))
plt.title('Student Marks')
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.bar(df['name'], df['marks'], color = 'brown')
```

```python
plt.figure(figsize=(16,9))
plt.title('Student Marks')
plt.xlabel('Student Name')
plt.bar(df['name'], df['marks'], color = 'gray', 
        width=0.5, align='edge', label = 'marks')
plt.bar(df['name'], df['age'], color = 'cyan', 
        width=0.5, align='center', label = 'age')
plt.legend()
```

```python
plt.figure(figsize=(16,9))
plt.title('Student Marks')
plt.ylabel('Student Name')
plt.xlabel('Marks')
plt.barh(df['name'], df['marks'], color = 'navy')
```

```python
plt.figure(figsize=(16,9))
plt.title('Student Marks')
plt.xlabel('Student Name')
plt.barh(df['name'], df['marks'], color = 'gray', 
        height=0.5, align='edge', label = 'marks')
plt.barh(df['name'], df['age'], color = 'cyan', 
        height=0.5, align='center', label = 'age')
plt.legend();
```
