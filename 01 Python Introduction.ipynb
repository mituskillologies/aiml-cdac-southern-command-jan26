#### Class and Object

```python
class Person:                        # Class definition
    def __init__(self, n, a):        # Constructor function 
        self.name = n                # name & age -> class variables
        self.age = a 
    def show(self):                  # Function of class
        print('Values are:', self.name, self.age)
    def isEligible(self):
        if self.age >= 18:
            return True
        else:
            return False
```

```python
# Create the object
p1 = Person("Ajay", 23)
```

```python
print(p1.name)
```

```python
p1.age
```

```python
p1.show()
```

```python
p2 = Person("Priya", 16)
```

```python
p2.show()
```

```python
p1.isEligible()
```

```python
p2.isEligible()
```

```python
p3 = Person("Ram", 34)
p3.name = "Raj"
p3.show()
p3.isEligible()
```
