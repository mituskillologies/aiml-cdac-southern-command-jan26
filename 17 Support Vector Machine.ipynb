#### Import the libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

#### Load the data

```python
df = pd.read_csv('datasets/wisc_bc_data.csv')
```

```python
df
```

#### Separate the input and output variables

```python
# input variables
x = df.drop(['id','diagnosis'], axis = 1)

# output variable
y = df['diagnosis']
```

```python
x.shape
```

```python
x.describe()
```

```python
y.value_counts()
```

```python
plt.pie(y.value_counts(), labels=y.value_counts().index, autopct='%2.2f%%');
```

#### Cross Validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

#### Build the model

```python
from sklearn.svm import SVC
```

```python
svm = SVC(random_state=0, kernel='linear')
```

```python
svm.fit(x_train, y_train)
```

#### Performance Evaluation

```python
y_pred = svm.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
accuracy_score(y_test, y_pred)
```

#### Experiment with different kernels

```python
svm_rbf = SVC(random_state=0, kernel='rbf')
svm_rbf.fit(x_train, y_train)
y_pred_rbf = svm_rbf.predict(x_test)
accuracy_score(y_test, y_pred_rbf)
```

```python
svm_poly = SVC(random_state=0, kernel='poly')
svm_poly.fit(x_train, y_train)
y_pred_poly = svm_poly.predict(x_test)
accuracy_score(y_test, y_pred_poly)
```

```python
svm_sig = SVC(random_state=0, kernel='sigmoid')
svm_sig.fit(x_train, y_train)
y_pred_sig = svm_sig.predict(x_test)
accuracy_score(y_test, y_pred_sig)
```

```python
svm_poly = SVC(random_state=0, kernel='poly', degree=10)
svm_poly.fit(x_train, y_train)
y_pred_poly = svm_poly.predict(x_test)
accuracy_score(y_test, y_pred_poly)
```
