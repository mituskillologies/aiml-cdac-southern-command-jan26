```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5,5,0.2)
```

#### Sigmoid

```python
def sigmoid(z):
    val = 1 / (1 + np.exp(-z))
    return val
```

```python
sig_y = sigmoid(x)
```

```python
plt.plot(x, sig_y)
```

```python
sigmoid(3.56)
```

```python
sigmoid(-3.89)
```

#### ReLU

```python
def relu(z):
    if z > 0:
        return z 
    else:
        return 0
```

```python
rel_y = [relu(y) for y in x]
```

```python
plt.plot(x, rel_y)
```

#### Softmax

```python
def softmax(z):
    val = np.exp(z) / np.sum(np.exp(z))
    return val
```

```python
x = [6,3,4,2]
```

```python
softmax(x)
```

```python
sum(softmax(x))
```

#### Hyperbolic Tangent

```python
def tanh(z):
    val = (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))
    return val
```

```python
x = np.arange(-5,5,0.2)
tan_y = tanh(x)
```

```python
plt.plot(x, tan_y)
```

```python
plt.grid()
plt.plot(x, sig_y)
plt.plot(x, tan_y)
```
