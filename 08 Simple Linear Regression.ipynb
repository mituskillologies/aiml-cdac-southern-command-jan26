```python
import numpy as np
```

```python
# Define input variable -> values of height in cms 
x = np.array([[151],[174],[138],[186],[128],[136],[179],[163],[152],[131]])

# Define output variable -> values of weight in kgs
y = np.array([63, 81, 56, 91, 47, 57, 76, 72, 62, 48])
```

```python
x
```

```python
import matplotlib.pyplot as plt 
plt.scatter(x, y)
```

```python
np.corrcoef(x.reshape(10), y)[0,1]
```

```python
# import the class
from sklearn.linear_model import LinearRegression
```

```python
# Create the object 
regressor = LinearRegression()
```

```python
# Train the model 
regressor.fit(x, y)
```

```python
regressor.predict([[180]])
```
