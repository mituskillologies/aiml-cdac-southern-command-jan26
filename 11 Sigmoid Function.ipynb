```python
def sigmoid(y):
    val = 1 / (1 + np.exp(-y))
    return val
```

```python
x = np.arange(-10,10,0.25)
```

```python
sig_y = sigmoid(x)
```

```python
sig_y
```

```python
plt.plot(x, sig_y)
```
